```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Drop unnamed column

```python
df=df.drop(['Unnamed: 0'],axis=1)
print(df.head)
```

# Gender distribution

```python
plt.figure(figsize=(6,6))
ax = sns.countplot(data=df,x='Gender')
ax.set_title('Gender Distribution')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.bar_label(ax.containers[0])
plt.show()
```

From the above chart we can analysed that number of females in the data is more than the number of males

```python
gb = df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
sns.heatmap(gb , annot=True)
plt.title('Parent Education vs Score')
plt.show()
```

From the above we can conclude that the education of parents have a good impact on the students scores
